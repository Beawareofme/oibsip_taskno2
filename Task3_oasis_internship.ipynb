# **Importing the Dependencies**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics 
```

# **Data Collection and Processing**

# **Encoding the Categorical Data**

```python
car_dataset.replace({'fueltype':{'gas':0,'diesel':1}},inplace=True)
car_dataset.replace({'drivewheel':{'fwd':0,'rwd':1,'4wd':2}},inplace=True)
car_dataset.replace({'cylindernumber':{'four':4,'six':6,'five':5,'eight':8,'two':2,'three':3,'twelve':12}},inplace=True)
car_dataset.replace({'carbody':{'sedan':0,'hatchback':1,'wagon':2,'hardtop':3,'convertible':4}},inplace=True)
```

```python
car_dataset
```

```python
# Splitting the data and

```

```python
X=car_dataset[['fueltype','carbody','drivewheel','wheelbase','cylindernumber','enginesize','horsepower']]
Y=car_dataset['price']
```

```python
X
```

```python
Y
```

# **Splitting Training and Test data**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

**1. Linear Regression model**

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,Y_train)
```

**Model Evaluation(on training data)**

```python
training_data_prediction = reg.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared Error :  ", error_score)
```

**Visualize the actual prices and Predicted prices(on training data)**

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price VS Predicted Price")
plt.show()
```

**Model Evaluation(on testing data)**

```python
testing_data_prediction = reg.predict(X_test)
error_score = metrics.r2_score(Y_test,testing_data_prediction)
print("R squared Error :  ", error_score)
```

**Visualize the actual prices and Predicted prices(on testing data)**

```python
plt.scatter(Y_test,testing_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price VS Predicted Price")
plt.show()
```

**2. Lasso Regression**

```python
las = Lasso()
```

```python
las.fit(X_train,Y_train)
```

**Model Evaluation(on training data)**

```python
training_data_prediction = las.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared Error :  ", error_score)
```

```python
testing_data_prediction = las.predict(X_test)
error_score = metrics.r2_score(Y_test,testing_data_prediction)
print("R squared Error :  ", error_score)
```
